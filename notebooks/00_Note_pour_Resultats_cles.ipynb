# Variable age :
---
Cette distribution est typique d’une entreprise ayant une pyramide des âges centrée sur les salariés en milieu de carrière.

# Variable DistanceFromHome
---
Cette proximité généralisée peut limiter les contraintes liées au transport et influencer positivement la satisfaction au travail.
ou 
La population étudiée se caractérise par une forte proximité domicile–travail : plus de 70 % des employés résident à moins de 10 km.  
Cela peut réduire l’impact du temps de trajet sur la fatigue ou le turnover potentiel.


# NumCompaniesWorked
---
➡️ C’est un indicateur de stabilité professionnelle
➡️ Potentiellement lié à :

- la fidélité à l’employeur

- la probabilité d’attrition

- l’expérience diversifiée

- le profil de carrière (junior, polyvalent, instable, etc.)

**regroupements les plus pertinents en RH (et les plus utilisés dans les analyses People Analytics)**

0–1 entreprises → profils très stables ou débutants

2–3 entreprises → mobilité modérée

4–6 entreprises → mobilité élevée

7+ entreprises → forte instabilité ou carrière très variée
