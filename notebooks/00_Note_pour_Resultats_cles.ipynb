# Variable DistanceFromHome
---
Cette proximité généralisée peut limiter les contraintes liées au transport et influencer positivement la satisfaction au travail.
ou 
La population étudiée se caractérise par une forte proximité domicile–travail : plus de 70 % des employés résident à moins de 10 km.  
Cela peut réduire l’impact du temps de trajet sur la fatigue ou le turnover potentiel.
