# Synthèse transversale et recommandations

## 1. Cadre

### 1.1 Rappel de la problématique et des questions d’analyse


Quels sont les effets individuels et combinés des conditions de travail, des facteurs personnels et du niveau de satisfaction et d’implication sur l’attrition des salariés ?


### Questions d’analyse

**1. Conditions personnelles**  
Quel est l’impact des conditions de vie personnelles sur le départ des collaborateurs ?

**2. Conditions de travail**  
Dans quelle mesure les conditions de travail influencent-elles l’attrition ?

**3. Satisfaction et implication**  
Comment le niveau de satisfaction et d’implication contribue-t-il à l’attrition ?  
Les conditions de travail et les facteurs personnels influencent-ils le niveau de satisfaction et d’implication ?

### Méthodologie de synthèse 

La synthese reprend les résultats observé lors de l'analyse exploratoire mené  en ammont sur les trois axes thématiques : **Axe 1 conditions personnelles**, **Axe 2 conditions de travail** et **l'Axe 3 satisfaction et implication**.

L'EDA s'est appuyé d'une analyse descriptive progressive des variables (individuelle, combiné à l'attrition, combiné entre elles) via des tableaux de proportion et de moyenne.

L'analyse de l'attrition s'est basé sur le taux global (environs 16%) comme référence pour les analyses.

Les résultats doivent être interprétés comme des indicateurs de relation et non comme des preuves de causalité.
Toute décision opérationnelle nécessiterait des analyses complémentaires de type causal.


##  2. Synthèse transversale des axes

### Qui part ? (Axe 1)

L' analyse mettent en évidence des profils pour lesquels les variables observés peuvent s’accompagner d’une probabilité relative de départ plus élevée, sans préjuger des trajectoires individuelles.

Les employés de moin de 25 ans et ayant un statut marital célibataire
Les employés ayant une mobilité professionnelle élevé et étant dans une tranches d'ages plus élevés
Les employés ayant le statut marital célibataire.

****
L’attrition est principalement associée à des phases de transition des parcours de vie et de carrière, notamment les débuts de carrière ou les situations de mobilité professionnelle, plutôt qu’à un âge ou un statut isolé.
Ces situations concernent plus fréquemment les salariés jeunes, célibataires ou ayant connu plusieurs expériences professionnelles.

### Dans quelles conditions ? (Axe 2)

L'analyse met en évidence des dynamique de départ selon les configurations organisationnelle observé en particulier :

- les salariés en début de parcours professionnel ( ancienneté dans l'entreprise, le poste, de la durée de collaboration avec le manager et l'experience professionnelle total), dans les premieres années.

- Le niveau de rémunération les plus faibles apparaissent comme un élément influeancant.

- Certaines fonction sont plus exposés notément les commercial, les rh et technician de laboratoire

- les heures supplémentaire peuvent joué un role mais combinés à d'autre facteur organisationnelle plus larges.

****

Les départs s’inscrivent majoritairement dans des contextes organisationnels marqués par une faible stabilisation professionnelle, en particulier lors des premières années dans l’entreprise, dans le poste ou dans la relation managériale.
La rémunération faible et certaines fonctions spécifiques renforcent cette exposition, tandis que des facteurs comme les heures supplémentaires jouent un rôle secondaire et contextuel.


### Pourquoi subjectivement ? (Axe 3 )

Le niveau de satisfaction et d'implication apparaissent comme une dimension central associé à l'attrition : Un niveau faible tend à l'augmentation de la décision de départs.

L'analyse croisé avec les deux autres axes met en évidences des dynamique organisationnelle relativement stable, tout en revélant quelques marqueur différenciés (pour les trajectoires longues ). Les perceptions de satisfactions ou d'implication varies peu selon les conditions observés, 



****

La satisfaction et l’engagement apparaissent comme des médiateurs clés entre les conditions objectives et la décision de départ.

Cependant, ces perceptions restent globalement stables selon de nombreuses conditions observées, ce qui suggère que la décision de départ ne repose pas uniquement sur des facteurs subjectifs immédiats, mais sur des dynamiques plus complexes et cumulatives.
