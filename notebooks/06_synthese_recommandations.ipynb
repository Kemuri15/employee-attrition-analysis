# Synthèse transversale et recommandations

---

## 1. Cadre

---

### 1.1 Rappel de la problématique et des questions d'analyse


Quels sont les effets individuels et combinés des conditions de travail, des facteurs personnels et du niveau de satisfaction et d'implication sur l'attrition des salariés ?

**Questions d'analyse**

- **1. Conditions personnelles**  
Quel est l'impact des conditions de vie personnelles sur le départ des collaborateurs ?

- **2. Conditions de travail**  
Dans quelle mesure les conditions de travail influencent-elles l'attrition ?

- **3. Satisfaction et implication**  
Comment le niveau de satisfaction et d'implication contribue-t-il à l'attrition ?  
Les conditions de travail et les facteurs personnels influencent-ils ces dimensions subjectives ?

### 1.2 Méthodologie de synthèse 

Cette synthèse s'appuie sur les résultats issus de l'analyse exploratoire menée selon trois axes thématiques :
**Axe 1 - Conditions personnelles**, **Axe 2 - Conditions de travail**, **Axe 3 - Satisfaction et implication.**

L'approche repose sur une analyse descriptive progressive :

- Analyse univariée des variables,
- comparaison avec l'attrition,
- exploration des relations internes entre variables.

Les analyses d'attrition prennent comme référence le taux global observé (≈ 16 %) afin d'identifier des situations de sur ou sous-exposition.

Les résultats doivent être interprétés comme des indicateurs de relation et non comme des preuves de causalité.
Toute décision opérationnelle nécessiterait des analyses complémentaires de type explicatif ou causal.

---

##  2. Synthèse transversale des axes

---

### 2.1 Qui part ? (Axe 1 - Conditions personnelles)


L'attrition apparaît davantage liée à des phases de transition dans le parcours de vie et de carrière qu'à des caractéristiques personnelles isolées telles que le statut marital.

Les départs sont plus fréquents :

- Chez les salariés jeunes, en début de parcours professionnel,
- chez certains profils plus expérimentés présentant une mobilité professionnelle élevée.

Ces résultats suggèrent que ce sont les dynamiques de parcours (entrée dans la vie professionnelle, changements de trajectoire) qui structurent davantage l'attrition, que les situations personnelles en tant que telles.

Ces éléments peuvent être interprétés comme des phases de transition, sans que celles-ci soient directement mesurées.

La distance domicile-travail apparaît comme un facteur contributif, mais secondaire, dont l'effet reste limité lorsqu'il est considéré isolément.

---


### 2.2 Dans quelles conditions ? (Axe 2 - Conditions de travail)


Les départs sont majoritairement associés à des configurations organisationnelles peu stabilisées, en particulier lors des phases d'intégration ou d'ajustement.

Les premières années dans l'entreprise, dans le poste ou dans la relation managériale constituent des zones de vulnérabilité accrues.

La rémunération joue également un rôle structurant :
les salariés aux niveaux de salaire les plus faibles présentent des taux d'attrition plus élevés, tout comme certaines fonctions spécifiques, notamment commerciales.

Les heures supplémentaires apparaissent fortement associées à l’attrition, avec un taux de départ nettement supérieur chez les salariés concernés. Toutefois, cette pratique ne concerne qu’une partie de l’effectif et doit être analysée en combinaison avec d’autres facteurs organisationnels.


### 2.3 Pourquoi subjectivement ?(Axe 3 - Satisfaction et implication)

La satisfaction et l'implication apparaissent comme des dimensions centrales dans la décision de départ.

Les analyses montrent que des niveaux faibles de satisfaction, d'engagement ou d'équilibre vie professionnelle / vie personnelle sont systématiquement associés à des taux d'attrition plus élevés.

Toutefois, le croisement avec les axes personnels et organisationnels met en évidence des dynamiques relativement stables :
les perceptions de satisfaction et d'implication varient peu selon les conditions observées.

Ces résultats suggèrent que la satisfaction et l'engagement ne constituent pas des déclencheurs immédiats et isolés du départ, mais plutôt des mécanismes intermédiaires, s'inscrivant dans une dynamique progressive et cumulative des conditions de travail et de trajectoires individuelles.


### 2.4 Lecture transversale : 


L'attrition ne résulte pas d'un facteur unique, mais de la combinaison :

- De phases de transition du parcours de vie (jeune âge),
- de situations professionnelles encore peu stabilisées (faible ancienneté),
- de conditions organisationnelles spécifiques,
- et de perceptions subjectives progressivement dégradées.

Cette lecture transversale met en évidence des trajectoires de départ construites dans le temps, plutôt qu'un phénomène ponctuel ou réactionnel, invitant à appréhender l'attrition comme un processus et non comme un événement isolé.

##  3. Leviers actionnables

La réponse à l'attrition combine des leviers opérationnels pour agir à court terme et des leviers structurels pour traiter les dynamiques de fond.

---

### 3.1 Les leviers opérationnels :


- ### **Renforcer l'onboarding managérial et proposer du mentorat entre salariés**

Ce levier vise à sécuriser les dynamiques de parcours des salariés dès les premières phases de leur intégration.
Ce levier cible prioritairement les salariés en début de parcours et lors des changements de poste.
Les résultats observés mettent en évidence une population plus exposée au risque d'attrition, notamment lors des premières années d'ancienneté, des premières prises de poste et des premières relations managériales.

Le renforcement de l'onboarding existant, combiné au développement de dispositifs de mentorat entre salariés, permettrait de mieux accompagner cette phase critique et de réduire les risques de décrochage précoce.

- ### **Accompagnement managérial**

Ce levier vise à sécuriser les premières relations manager-collaborateur, qui influencent fortement la perception du travail, l'engagement et la satisfaction globale.
Les relations managériales récentes constituent une zone de vulnérabilité, tant du côté du collaborateur que du manager, et peuvent contribuer à une dégradation progressive de l'engagement.

Un accompagnement ciblé permettrait de stabiliser la relation managériale et de prévenir les situations à risque.

- ### **Vigilance sur le recours aux heures supplémentaires**

Ce levier vise à limiter le cumul de facteurs susceptibles d’aggraver des situations déjà fragilisées.

Dans la continuité des résultats relatifs aux conditions de travail, les heures supplémentaires apparaissent comme un facteur aggravant, associé à un taux d’attrition significativement plus élevé. Elles semblent toutefois agir en interaction avec d’autres facteurs de fragilité (ancienneté faible, poste sous tension, phase de transition), plutôt que comme un déclencheur isolé.

Une vigilance accrue sur ce sujet permettrait de prévenir les effets cumulatifs pouvant conduire à une décision de départ 

- ### **Reconnaissance non salariale** 

Ce levier immatériel vise à agir sur la perception et le ressenti des salariés, en dehors des mécanismes de reconnaissance salariale.
Les analyses relatives à la satisfaction et à l'engagement montrent que ces dimensions jouent un rôle de mécanisme intermédiaire dans l'attrition.

Des actions de reconnaissance managériale ou d'autres dispositifs non monétaires pourraient ainsi renforcer la perception de valeur et contribuer à la fidélisation.


Ces leviers opérationnels nécessitent d'être soutenus par des dispositifs structurels afin d'assurer leur pérennité.



### 3.2 Les leviers structurels


- **Politique d'intégration, de mentorat et de suivi des nouveaux salariés et des mobilités internes**

Une politique d'intégration et de mentorat vise à sécuriser les premières années de parcours en offrant un cadre rassurant, un accompagnement personnalisé et un soutien relationnel complémentaire au management. Elle contribue ainsi à renforcer l'engagement durable des collaborateurs et à réduire l'attrition des populations en début de parcours.

Dans une logique de pérennisation des leviers opérationnels, ce levier contribuera à la sécurisation des dynamiques de parcours lors des premières années, tant sur les plans professionnel, relationnel et d'intégration à la culture d'entreprise.
Il participe également à l'amélioration de l'expérience collaborateur et au renforcement de la cohésion et de la culture d'entreprise, en favorisant les liens intergénérationnels et intermétiers.


- **Modèle managérial et plan de formation** 

Afin d'ancrer durablement les pratiques managériales identifiées, ce levier vise à définir un cadre managérial commun au sein de l'organisation (valeurs, posture, pratiques clés) et l'ancrer durablement par la formation.

Il permettrait un alignement entre la stratégie de réduction des risques d'attrition, les pratiques managériales et la culture de l'organisation.


- **Politique salariale ciblée sur les bas salaires** 
 
Au regard du rôle structurant de la rémunération sur l'attrition, ce levier vise à renforcer la rétention des populations les plus exposées au risque de départ, en particulier les salariés à bas salaire.
La rémunération constitue un facteur d'influence majeur du turnover.

Une politique salariale ciblée permettrait non seulement de renforcer la rétention, mais également d'assurer une meilleure cohérence des pratiques salariales, d'anticiper les risques sociaux et de soutenir durablement la satisfaction et l'engagement.

## 4. Configurations de risque

Cette section vise à présenter des configurations de risque associées à une exposition plus élevée à l'attrition.

Il ne s'agit pas de profils prédictifs, mais de configurations observées lors de l'analyse exploratoire.
Elles sont construites à partir de la combinaison de facteurs identifiés comme influençant ou contribuant à l'attrition.

Les trois configurations présentées combinent des facteurs de parcours, organisationnels et subjectifs.

---

### **4.1 Les jeunes salariés en début de parcours**


### **Combinaison de facteurs**

- Jeunes salariés
- Première prise de poste
- Faible ancienneté
- Niveau de salaire faible
- Niveau de satisfaction et d'engagement faible

### **Logique de risque**

Cette configuration présente un cumul de facteurs associés à un taux d'attrition élevé.
La phase de début de parcours correspond à une période de transition marquée par une faible stabilisation organisationnelle. Elle constitue une zone de vulnérabilité.

La combinaison d'une rémunération faible et de niveaux réduits de satisfaction et d'engagement renforce cette exposition au risque de départ.
Cette configuration correspond à une phase d'apprentissage et d'ajustement, où les attentes et la réalité du poste peuvent diverger.

### **Leviers associés**

- Onboarding renforcé pour améliorer la stabilisation organisationnelle
- Accompagnement managérial pour sécuriser la relation manager-collaborateur
- Reconnaissance non salariale afin d'agir sur la satisfaction et l'engagement (leviers subjectifs)

---

### **4.2 Salariés expérimentés en phase de transition**

### **Combinaison de facteurs**

- Salariés expérimentés
- Prise de poste récente
- Collaboration récente avec le manager
- Heures supplémentaires
- Distance domicile-travail élevée
- Niveau de satisfaction faible

### **Logique de risque**

Cette configuration se caractérise par une exposition accrue au risque de départ.
La phase de transition professionnelle, combinée à une faible stabilisation organisationnelle, constitue une zone de vulnérabilité.

La présence de facteurs contributifs supplémentaires, tels que les heures supplémentaires ou la distance domicile-travail (associées à un taux d’attrition plus élevé), peut accentuer une dégradation progressive du niveau de satisfaction. 
L'ensemble de ces éléments forme une configuration défavorable.

### **Leviers associés**

- Onboarding renforcé et mentorat pour améliorer la stabilisation organisationnelle
- Accompagnement managérial pour sécuriser la relation de travail
- Vigilance sur les heures supplémentaires afin de limiter les facteurs aggravants
- Reconnaissance non salariale pour soutenir la satisfaction et l'engagement (leviers subjectifs)

---

### **4.3 Salariés commerciaux en forte mobilité**

### **Combinaison de facteurs**

- Poste commercial
- Salariés dans une tranche d'âge élevée
- Forte mobilité professionnelle
- Heures supplémentaires
- Faible niveau de reconnaissance et d'engagement

### **Logique de risque**

Cette configuration est associée à une situation défavorable.
Les conditions de parcours spécifiques, telles que l'âge et la mobilité professionnelle, se combinent à des contraintes organisationnelles liées au poste et aux heures supplémentaires.

Un faible niveau d'engagement et de reconnaissance renforce cette zone de vulnérabilité et contribue à une exposition accrue au risque d'attrition.

### **Leviers associés**

- Vigilance sur les heures supplémentaires afin de limiter les facteurs aggravants
- Reconnaissance non salariale pour renforcer la satisfaction et l'engagement (leviers subjectifs)


Ces configurations ne sont pas exhaustives.
Elles visent à orienter les actions RH et peuvent évoluer dans le temps.

Ces configurations illustrent la nature multifactorielle de l'attrition et soulignent l'importance d'actions différenciées selon les phases de parcours et les contextes organisationnels.

## 5. Limites et perspectives
---

### 5.1 Limites de l'analyse

- **Limites liées à la méthode (EDA)**

Cette analyse repose sur une approche exploratoire et descriptive, basée sur l'observation de relations entre variables.
Elle ne permet pas d'établir de liens de causalité entre les facteurs étudiés et l'attrition.

De plus, l'absence de modélisation prédictive ne permet pas d'estimer l'impact relatif de chaque facteur ni de prédire le risque individuel de départ.

Une analyse complémentaire croisant les fonctions occupées avec d’autres variables (rémunération, ancienneté, heures supplémentaires) permettrait d’affiner la compréhension des dynamiques propres à certains métiers.

- **Limites liées aux variables**

L'analyse s'appuie sur un jeu de données contenant plusieurs variables auto-déclarées, notamment celles relatives à la satisfaction, à l'engagement et à l'équilibre vie professionnelle / vie personnelle. Ces indicateurs peuvent être influencés par des biais de perception ou de déclaration.

Par ailleurs, certaines dimensions importantes ne sont pas mesurées, telles que le climat social, la culture d'entreprise ou la qualité managériale réelle. Certaines variables RH structurantes sont également absentes, notamment la distinction entre turnover volontaire et involontaire.

Ces limites restreignent la capacité à appréhender de manière exhaustive les mécanismes sous-jacents à l'attrition et impliquent une lecture prudente des résultats.

- **Limites liées au contexte**

Cette analyse repose sur un jeu de données fictif représentant une entreprise dans un contexte spécifique. Les résultats observés sont donc fortement dépendants de ce cadre.

L'approche adoptée offre une vision à un instant donné et ne permet pas d'analyser l'évolution des comportements ou des facteurs de risque dans le temps.

Enfin, ces éléments limitent la généralisation des résultats à d'autres organisations ou contextes sectoriels.

---

### 5.2 Perspectives d'approfondissement

- **Perspectives analytiques**

Afin de prolonger cette analyse exploratoire, il serait pertinent de recourir à des méthodes de modélisation statistique, telles que la régression logistique, afin d'estimer l'effet combiné des facteurs identifiés sur l'attrition.

Des analyses multivariées intégrant des interactions entre variables ou des approches de segmentation permettraient également d'affiner la compréhension des profils et configurations de risque.

- **Perspectives données**

L'accès à des données longitudinales permettrait de compléter l'analyse par une approche temporelle, en étudiant l'évolution des parcours, de la satisfaction et des comportements de départ dans le temps.

Dans le cadre du présent projet, cette perspective reste toutefois limitée en raison du caractère fictif du jeu de données.

- **Perspectives RH / opérationnelles**

Sur le plan opérationnel, les leviers identifiés pourraient faire l'objet d'un suivi dans le temps à travers la mise en place d'indicateurs de pilotage RH.

Ce suivi permettrait d'évaluer l'impact des actions mises en œuvre, d'ajuster les dispositifs d'accompagnement et de renforcer une démarche proactive de prévention de l'attrition.



Ces limites et perspectives rappellent que l'attrition est un phénomène complexe, nécessitant des approches complémentaires et évolutives.

---


## 6. Conclusion générale


Ce projet avait pour objectif d'analyser les facteurs associés à l'attrition des salariés en croisant trois dimensions complémentaires : les conditions personnelles, les conditions de travail et les dimensions subjectives liées à la satisfaction et à l'implication.

L'analyse exploratoire met en évidence que l'attrition ne repose pas sur un facteur unique, mais sur la combinaison de dynamiques de parcours, de configurations organisationnelles et de perceptions individuelles. Les départs apparaissent particulièrement concentrés lors des phases de transition, notamment en début de carrière ou lors de prises de poste récentes, lorsque la stabilisation organisationnelle, relationnelle et salariale est encore fragile.

Les conditions de travail jouent un rôle structurant, en particulier l'ancienneté, l'expérience professionnelle, la rémunération et certaines fonctions spécifiques.D'autres facteurs, tels que les heures supplémentaires, apparaissent davantage comme des éléments contributifs que comme des déclencheurs directs. Les conditions personnelles permettent quant à elles de structurer des profils de salariés plus exposés au risque de départ. 

Toutefois, ces éléments ne suffisent pas à expliquer seuls les comportements de départ observés. Les dimensions subjectives-satisfaction, engagement et équilibre vie professionnelle / vie personnelle apparaissent comme des mécanismes intermédiaires majeurs, fortement associés à l'attrition, sans être directement déterminées par une variable organisationnelle unique.

Ainsi, l'attrition peut être abordée comme un processus progressif et multifactoriel, résultant de l'accumulation de facteurs de vulnérabilité plutôt que comme un événement soudain déclenché par une cause isolée. Cette lecture souligne l'importance d'une approche globale, combinant actions organisationnelles, managériales et attention portée au ressenti des salariés.

En ce sens, ce travail met en évidence des leviers RH concrets permettant d'agir en amont sur les situations à risque, tout en rappelant la nécessité d'analyses complémentaires pour approfondir les mécanismes causaux. 
Il constitue une base de réflexion pour orienter des politiques de prévention de l'attrition plus ciblées, différenciées selon les phases de parcours et les contextes organisationnels.
